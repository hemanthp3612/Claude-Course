```python

# Main code to interact with the API along with helper functions to manage the conversation

# The code starts by importing the load_dotenv function from the dotenv library, which is used to load environment variables from a .env file. 
# This allows the API key to be stored securely in a separate file and accessed in the code without hardcoding it.
from dotenv import load_dotenv

load_dotenv()

from anthropic import Anthropic

# The Anthropic class is imported from the anthropic library, which provides a client for interacting with the API. 
# An instance of the Anthropic class is created and stored in the client variable, which will be used to send requests to the API.
client = Anthropic()

from os import system

# Define the model to use
model = "claude-sonnet-4-6"

#Make a request to the API
# The messages parameter is a list of dictionaries, where each dictionary represents a message in the conversation. 
# Each message has a "role" (either "user" or "assistant") and "content" (the text of the message).
def add_user_message(messages, text):
    user_message =  {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

# The add_assistant_message function is a helper function that adds a message from the assistant to the conversation. 
# It takes the list of messages and the text of the assistant's response as input, creates a new message dictionary with the role "assistant", and appends it to the messages list.
def add_assistant_message(messages, text):
    assistant_message =  {
        "role": "assistant",
        "content": text
    }   
    messages.append(assistant_message)

# The chat function is the main function that interacts with the API. 
# It takes a list of messages and an optional system message as input, constructs the parameters for the API request, and sends the request to the API. The response from the API is then returned as text.
def chat(messages, system=None, temperature = 1.0):

    # The params dictionary is constructed with the model, maximum tokens, and messages. 
    # If a system message is provided, it is also added to the params dictionary.
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    # If a system message is provided, it is added to the params dictionary under the key "system". 
    # This allows the API to understand any specific instructions or context that should be applied to the conversation.
    if system:
        params["system"] = system
    
    # The client.messages.create method is called with the params dictionary to send the request to the API. 
    # The response from the API is stored in the message variable, and the text of the assistant's response is extracted and returned.
    message = client.messages.create(**params)   
    return message.content[0].text
```

```python
# The code then initializes an empty list of messages and adds a user message asking for the capital of France.

messages = []
add_user_message(messages, "What is the capital of France? Answer in single line")
response = chat(messages)
print(response)
```

```python
# The code then adds the assistant's response to the messages list and continues the conversation by asking another question about the population of France.
add_assistant_message(messages , response)
add_user_message(messages, "What is the population of France? Answer in single line")
response = chat(messages)
print(response)
```

```python
# Make a mini chatbot with Multiple turns of conversation

messages = []

while True:
    user_input = input(">> ")

    print(">> ", user_input)
    add_user_message(messages, user_input)
    response = chat(messages)
    
    print("-------")
    print(response)
    print("-------")    
    add_assistant_message(messages , response)
    
```

```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"

messages = []
messages.append({
    "role": "user",
    "content": "Generate a very short EventBridge rule as JSON. Wrap it in <json> and </json> tags."
})

response = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stop_sequences=["</json>"]  # stops when closing tag is hit
)

text = response.content[0].text
# Strip the opening tag
json_text = text.replace("<json>", "").strip()

print(json_text)
print("Stop reason:", response.stop_reason)  # should print 'stop_sequence'
```
